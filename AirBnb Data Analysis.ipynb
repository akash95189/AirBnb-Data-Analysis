## Analysing Price Distribution

```python
sns.histplot(df, x='price')
```

```python
sns.boxplot(df, x='price')
```

```python
df1 = df[df['price'] < 1500]
```

```python
sns.histplot(df1, x='price', bins=100)
```

```python
sns.boxplot(df1, x='price')
```

```python
df1.columns
```

```python
plt.figure(figsize=(4,2))
sns.histplot(df1, x='availability_365')
plt.show()
```

## Making Dataset Concise for better Visualisation

```python
df1.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df1['price_per_bed']=df1['price']/df1['beds']
```

```python
df1.head()
```

```python
df1.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
df1.columns
```

## Analysing Type of Room with Price

```python
plt.figure(figsize=(10,8))
sns.barplot(df1, x='neighbourhood_group', y='price', hue='room_type')
plt.title("Type of Room in different Locality with Price")
plt.xlabel("Locality")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(df1, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title("Locality and Review Analysis")
plt.xlabel("Reviews")
plt.ylabel("Price")
plt.show()
```

```python
df1.columns
```

## Comparing Price, Minimum Nights,Number of Reviews and Availability of Rooms based on Type of Rooms

```python
plt.figure(figsize=(10,8))
sns.pairplot(df1, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
plt.show()
```

## Geographical Distribution of AirBnb Listing

```python
plt.figure(figsize=(10,8))
sns.scatterplot(df1, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.xlabel("Longitude")
plt.ylabel("latitude")
plt.show()
```

## Detemining Co-relation between Latitude, Longitude, Minimum Nights, Number of Reviews, Reviews per Month, Availability of Rooms and Number of Beds using Heatmap

```python
df1.columns
```

```python
corr = df1[['latitude','longitude', 'minimum_nights','number_of_reviews', 'reviews_per_month','availability_365', 'beds']].corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True)
plt.title("Co-relation Heatmap")
plt.show()
```
